```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
auc_response_time = pickle.load(open( "auc_response_time.bin", "rb" ))
auc_conf = pickle.load(open( "auc_confidence.bin", "rb" ))
```

```python
auc_response_time
```

```python
auc_conf
```

```python
from scipy import stats
x = []
y = []
for k in auc_conf:
    x.append(auc_conf[k])
    y.append(auc_response_time[k])
    
    

plt.figure(figsize=(5,5))

# Scatter points
plt.title('AUC Scatter-Plot')
plt.xlabel('AUC Confidence')
plt.ylabel('AUC Response Time')
plt.axis(xmin=0.4, xmax=1, ymin=0.4, ymax=1)
plt.scatter(x,y,50,alpha=0.3, c="black",label="")
plt.plot([0, 1], [0, 1], c='black',label="Diagonal")
# Trendline
slope, intercept, r_value, _, _ = stats.linregress(x,y)
x_fit = [i/10 for i in range(4,11)]
y_fit = [intercept + slope*i for i in x_fit]
plt.plot(x_fit, y_fit, 'r', label="Trendline\nr = "+str(round(r_value,2)))
l = plt.legend(loc=4)
```
